Q: A description of the problem and a discussion of the background.

A:

My client want to open an office in Berlin, Germany, and he want to find a suitable place to rent.  

For a convenient and efficient working environment, the office should be close to food and cafe. The client also want to access to a specific district where he could have many choices.

Q: A description of the data and how it will be used to solve the problem.

A:
Since the client wants to rent an office in Berlin, Germany. Our data will be generated by querying on Foursquare. 

Data utilization is shown as followed:

a. we used "office" as key word to gain geological information about office in Berlin. 

b. we filted sites from the category "Office".

c. we found surrounding venues of selected offices within 500 meters and summerised the frequency of occurrence of each 
category. We also printed the top 5 most common venues of each office site.

e. we run k-means to cluster the offices into 5 clusters.

f. we examine each cluster and determine the perfect district to rent the ideal office.
